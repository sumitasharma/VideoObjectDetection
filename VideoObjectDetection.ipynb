# Use YOLOv3 pretrained model to detect and output video with recognized objects.

Set the path variable to the data set folder path on Google Drive

```
from imageai.Detection import VideoObjectDetection
import os

# Set path to the drive folder where video data set is stored
path = '/content/gdrive/My Drive/Colab Notebooks/VideoDataSet'
execution_path = os.getcwd()

detector = VideoObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath( os.path.join(execution_path , path + "/yolo.h5"))
detector.loadModel()

video_path = detector.detectObjectsFromVideo(input_file_path=os.path.join( execution_path, path + "/VID_20181117_133018.mp4"),
                                output_file_path=os.path.join(execution_path, path + "/VID_20181117_133018_1")
                                , frames_per_second=29, log_progress=True)
print(video_path)
```
